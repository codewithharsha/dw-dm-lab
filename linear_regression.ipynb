```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error,r2_score
diabetes_X,diabetes_Y = datasets.load_diabetes(return_X_y=True)
#diabetes_X
```

```python
diabetes_X = diabetes_X[:,np.newaxis,2]
#diabetes_Y
```

```python
#splitting into training & testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
diabetes_Y_train = diabetes_Y[:-20]
diabetes_Y_test = diabetes_Y[-20:]
```

```python
#visualizing the data
plt.scatter(diabetes_X_train,diabetes_Y_train)
plt.xlabel("x axis")
plt.ylabel("Y axis")
plt.show()
```

```python
#creating instance of linera regression
model = linear_model.LinearRegression()
#fitting the data
model.fit(diabetes_X_train,diabetes_Y_train)
```

```python
#predction of data
diabetes_Y_pred = model.predict(diabetes_X_test)
print(diabetes_Y_pred)
```

```python
r2_scor= r2_score(diabetes_Y_test,diabetes_Y_pred)
print(int(r2_scor*100),'%')
```

```python
plt.scatter(diabetes_X_test,diabetes_Y_test,color='b')
plt.plot(diabetes_X_test,diabetes_Y_pred,color='red')
plt.show()
```

```python
plt.scatter(diabetes_X_test,diabetes_Y_test,color='black')
plt.plot(diabetes_X_test,diabetes_Y_pred,color='red',linewidth=2,marker='*')
```

```python
plt.scatter(diabetes_X_test,diabetes_Y_test,color='black')
plt.plot(diabetes_X_test,diabetes_Y_pred,color='blue',linewidth=2,marker='*')
plt.grid()
```

```python
print("Mean Squared Error=%.2f"%r2_score(diabetes_Y_test,diabetes_Y_pred))
```
